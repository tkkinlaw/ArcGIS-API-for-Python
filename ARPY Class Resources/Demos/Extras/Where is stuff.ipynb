```python
from arcgis.gis import GIS
from arcgis.apps.itemgraph import create_dependency_graph

import networkx as nx
import matplotlib.pyplot as plt
```

```python
gis = GIS('Pro')
```

```python
items = gis.users.me.items(max_items=300)
```

```python
maps = []
for item in items:
    if item.type == "Web Map":
        print(item.title)
        maps.append(item)

        graph2 = create_dependency_graph(gis,maps,False)
        gItems = graph2.all_items()

        labels = {}
        colorMap = []

        for wItem in gItems:
            labels[wItem.id] = wItem.item.type

            if wItem.item.type == 'Web Map':
                colorMap.append('red')
            if wItem.item.type == 'Feature Service':
                colorMap.append('blue')
            if wItem.item.type == 'Service Definition':
                colorMap.append('pink')
            elif wItem.item.type == 'Web Experience':
                colorMap.append('green')
            else:
                pass

        nx.draw_networkx(graph2, with_labels=True, node_size=1500, node_color=colorMap, labels=labels, font_size=15,font_weight="bold", arrowsize=20)



        plt.tight_layout()
        #plt.figure().set_figwidth(15)

        plt.savefig(f"C:\DemoData\{wItem.item.title}.png",dpi=1500)

        print(gItems)
```
