```python
from arcgis.gis import GIS
from getpass import getpass
```

```python
url = r'https://www.arcgis.com/'
targetFolder = 'ARPY Exercises'
```

```python
gisTarget = GIS(url, 'tkinlaw', '10June2019!')
gisTarget
```

```python
gisSource = GIS(url, 'auto120_ts', getpass())
gisSource
```

```python
contentList = gisSource.content.search(query = f'type: feature service, owner:{gisSource.users.me.username}', max_items=100)
print(f"Cloning {len(contentList)} items")
gisTarget.content.clone_items(items=contentList, folder=targetFolder)
print("Done")
```

```python
webmapList = gisSource.content.search(query = f'type: web map, owner:{gisSource.users.me.username}', max_items=100)
print(f"Cloning {len(webmapList)} items")
gisTarget.content.clone_items(items=webmapList, folder=targetFolder)
print("Done")
```
