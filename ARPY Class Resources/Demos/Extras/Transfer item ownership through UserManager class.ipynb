# Reassign owwnership of items in ArcGIS Online

```python
from arcgis.gis import GIS
from getpass import getpass
```

```python
gis = GIS('https://www.arcgis.com', 'tkinlaw', getpass())
```

```python
sourceUser = gis.users.get('JaneDoe0')
targetUser = gis.users.get('JohnDoe748')
```

## Use Content Manager search to create a list of items to identify scope of changes

```python
# search returns a list of items
contentList = gis.content.search(query=f'owner:{sourceUser.username}')
print(f"transferring {len(contentList)} from {sourceUser.username} to {targetUser.username}")
```

## Use the User Manager items method to transfer ownership. 
This method returns a `generator`. Count the output to verify it's the same as `search` results. You can use the `reassign_to` method when accessing the `items` from the `Content Manager`. This is a more scalable way to doing this (in case you did this for multiple users).


```python
count = 0
for item in sourceUser.items():
    item.reassign_to(targetUser)
    count += 1
print(f"{count} items reassigned.")
```
