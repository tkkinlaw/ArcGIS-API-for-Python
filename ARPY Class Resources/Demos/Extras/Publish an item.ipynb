```python
from arcgis.gis import GIS
```

```python
gis = GIS('Pro')
itemDetails = {'title':'DC Public Schools py',
            'description':'Demo publishing from python',
            'tags':'arcgis, python, demo'}
data = r"C:\DemoData\FedGIS\Open_Data_DC\DC_Public_Schools.zip"
```

```python
folder = gis.content.folders.get()
itemZip = folder.add(itemDetails, data).result()
item = itemZip.publish()
```
