```python
from arcgis.gis import GIS
from getpass import getpass
from datetime import datetime as dt
from datetime import timedelta, timezone
import pandas as pd
```

```python
gis = GIS('https://www.arcgis.com', 'tkinlaw', getpass())
```

```python
allUsers = gis.users.search('')
print(len(allUsers))
```

```python
data = []
for user in allUsers:
    lastLogin = dt.fromtimestamp(user.lastLogin/1000)
    print(user.username, lastLogin, type(lastLogin))
    dict = {'username':user.username,
            'lastLogin':dt.fromtimestamp(user.lastLogin/1000, tz=timezone.utc)}
    data.append(dict)
```

```python
nowUTC = dt.now(timezone.utc)
cutoffUTC = nowUTC - timedelta(days=7)
```

```python
df = pd.DataFrame(data)
df_sorted = df.sort_values(by="lastLogin", ascending=True)
df_sorted
```
