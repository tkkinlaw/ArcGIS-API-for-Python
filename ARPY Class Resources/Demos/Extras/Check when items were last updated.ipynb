# Check when items were last updated

```python
from arcgis.gis import GIS
from getpass import getpass
from datetime import datetime as dt
import pandas as pd
```

```python
gis = GIS('https://www.arcgis.com', 'tkinlaw', getpass())
gis
```

```python
allContent = gis.content.search('', max_items=500, sort_field = 'type', sort_order="desc")
print(len(allContent))
```

## Order content by date modified

```python
rows = []
for item in allContent:
    rows.append({
        "title": item.title,
        "item": item,
        "type": item.type,
        "dateModified": item.modified, #don't convert to a readble date here (a string) because it will affect
        "readableModifiedDate": dt.fromtimestamp(item.modified/1000).strftime('%D') 
    })
df = pd.DataFrame(rows)
df_sorted = df.sort_values(by="dateModified", ascending=True).head(10)
df_sorted
```
