```python
from arcgis.gis import GIS
```

```python
gis = GIS("Pro")
gis
```

```python
newField = {"fields": [{"name": "tree_condition",
                        "type": "esriFieldTypeString",
                        "alias": "Tree Condition",
                        "nullable": True,
                        "editable": True,
                        "domain": None
                       }]}
```

```python
myContent = gis.users.me.items()
for item in myContent:
    if "TreeSurvey" in item.title and item.type == "Feature Service":
        print(item.title)
        flManager = item.layers[0].manager
        flManager.add_to_definition(newField)
print("Process complete")
```

```python
newDomainUpdate = {"fields": [{
    "name": "tree_condition",
    "domain": {
        "type": "codedValue",
        "name": "tree_condition_domain",
        "codedValues": [
            {"name": "Good", "code": "Good"},
            {"name": "Fair", "code": "Fair"},
            {"name": "Poor", "code": "Poor"}
            ]}}]}
```

```python
myContent = gis.users.me.items()
for item in myContent:
    if "TreeSurvey" in item.title and item.type == "Feature Service":
        print(item.title)
        flManager = item.layers[0].manager
        flManager.update_definition(newDomainUpdate)
print("Process complete")
```
